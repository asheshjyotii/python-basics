## Write a function named print_right that takes a string named text as a parameterand prints the string with enough leading spaces that the last letter of the string is inthe 40th column of the display

```python
def print_right(text):
    length = len(text)
    required = 40-length
    final = ' ' * required + text 
    print(f"->{final}")
    print(f"length : len(final)")
list = ['ahsesh is a good boy','lalpi is a bad girl', 'sunita is a bad girl']

for x in list :
    print_right(x)
```

## Write a function called triangle that takes a string and an integer and draws a trian‐gle with the given height, made up of copies of the string. Here’s an example of a tri‐angle with five levels using the string '@'

```python
def triangle(icon, height):
    for x in range(height):
        # print(icon*width)
        for y in range(x+1):
            print(icon, end = '')
        print("")

triangle('*', 50)
```

## Write a function called rectangle that takes a string and two integers and draws a rectangle with the given width and height, made up of copies of the string.

```python
def rectangle(icon, width, height):
    for x in range(height):
        print(icon*width)

rectangle('#', 40, 8)
```

## Class & object creation

```python
class Fire:
    def __init__(self, size, temp):
        self.size = size
        self.temp = temp
    def attr(self):
        print(f"The size of flame is : {self.size}")
        print(f"The temperature of flame is : {self.temp}")
```

```python
small = Fire(20, 300)
large = Fire(50, 500)


small.attr()
large.attr()
```

```python
class CloudAgent:

    API_KEY = "123" # calss attr
    credits = 30000
    
    def __init__(self, name, credit_req):
        self.name = name # instance attr
        self.credit_req = credit_req

    def token_usage(self):
        CloudAgent.credits = CloudAgent.credits-self.credit_req
        print(f"Agent '{self.name}' requires {self.credit_req} credits.")
        print(f"Agent '{self.name}' used {self.credit_req} credits.")
        print(f"The API KEY is {CloudAgent.API_KEY}")
        print(f"The available {CloudAgent.credits} are.")

    @classmethod
    def change_api_key(cls, new_api_key):
        cls.API_KEY = new_api_key
    
        

    
```

```python
agent_1 = CloudAgent("google_01", 10000)
agent_2 = CloudAgent("claude_01", 15000)

agent_1.token_usage()
agent_2.change_api_key("567")
agent_2.token_usage()
```
